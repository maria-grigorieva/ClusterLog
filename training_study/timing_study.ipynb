```python
import matplotlib
import matplotlib.pyplot as plt 
import numpy as np
from matplotlib.ticker import AutoMinorLocator
from datetime import datetime
```

```python
  
label1='Days'
label2='total_messages'
label3='cleaned_messages'
label8='number of clusters'
label4='cleaning_time'
label5='tokenization_time'
label6='training_time'
label7='vocabulary_size'

fields=[label1,label2,label3,label4,label5,label6,label7]
fields_df=[label2,label3,label4,label5,label6,label7]
```

```python
df=pd.read_csv('training_study.csv', names=fields_df)
```

```python
df.shape[0]
```

```python
df.reset_index(drop=True,inplace=True)
```

```python
df.head()
```

```python
days=range(0,df.shape[0])
```

```python
df_old=pd.read_csv('training_parameters_nik.csv', names=fields_df)
df_old.head()
```

```python
name1='Days'
name2='Total messages'
name3='Cleaned messages'
name8='Number of clusters'
name4='Cleaning time'
name5='Tokenization time'
name6='Training time'
name7='Vocabulary size'
```

```python
plt.figure(figsize=(10,7))
ax2=plt.subplot()
ax2.grid(axis='both', alpha=0.3,linewidth=1,linestyle='--',color='black')
#ax.xaxis.set_minor_locator(AutoMinorLocator())
#ax.yaxis.set_minor_locator(AutoMinorLocator())

ax2.plot(days,df[label4].values, 'rs',label='Nucleus '+name4,markersize=3)
ax2.plot(days,df_old[label4].iloc[:df.shape[0]],'gs',label='SOTA '+name4,markersize=3)
ax2.tick_params(axis="both", direction="in", length=4, width=2,labelsize=17)
ax2.ticklabel_format(axis='y',style='sci')

ax2.set_xlabel(label1,fontsize=17)
ax2.tick_params(axis="both", direction="in", length=8, width=2,labelsize=17)
ax2.set_ylabel(name4+' [s]',fontsize=17)
plt.legend(fontsize=17,loc='upper left',markerscale=2)
plt.title("Cleaning time: Nucleus VS SOTA",fontsize=17)

plt.savefig("modelmin_count2_default_cleanshortTrue/cleaning_timeNUCLEUS")
```
