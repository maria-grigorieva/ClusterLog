```python
from clusterlogs import pipeline
```

```python
import pandas as pd
```

```python
f = open('samples/syslog_fixed.log', 'r')
all = f.readlines()
f.close()
df = pd.DataFrame(all, columns=['message'])
print(df)
target = 'message'
```

```python
cluster = pipeline.Chain(df, target, model_name='models/syslog_new.model', mode='create',
                         add_placeholder=True, matching_accuracy=0.8,
                         clustering_type='ML', output_file='reports/syslog.html',
                         categorization=False,
                         generate_html_report=False)
```

```python
cluster.process()
```
